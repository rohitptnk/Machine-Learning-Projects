```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Study Materials /8th Sem/4. NLP/Assignments/Assignment_1_2025-main/hate/train.csv')

df
```

```python
texts = list(df['Sentence'])
texts[0:5]
```

```python
tags = list(df['Tag'])
tags[0:5]
```

```python
vectorizer = CountVectorizer(ngram_range=(1,3))

X_train = vectorizer.fit_transform(texts)
y_train = tags
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
df_val = df = pd.read_csv('/content/drive/MyDrive/Study Materials /8th Sem/4. NLP/Assignments/Assignment_1_2025-main/hate/val.csv')

df_val.head()
```

```python
X_test = vectorizer.transform(list(df_val['Sentence']))
y_test = list(df_val['Tag'])
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(accuracy)
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)
```

```python
df_humor = pd.read_csv('/content/drive/MyDrive/Study Materials /8th Sem/4. NLP/Assignments/Assignment_1_2025-main/humor/train.csv')

df_humor
```

```python
df_humor['Tag'].value_counts()
```

```python
df_humor_val = df = pd.read_csv('/content/drive/MyDrive/Study Materials /8th Sem/4. NLP/Assignments/Assignment_1_2025-main/humor/val.csv')
df_humor_val
```

```python
humor_tags = list(df_humor['Tag'])
humor_texts = list(df_humor['Sentence'])
```

```python
vectorizer = CountVectorizer(ngram_range=(1,3))

X_train_humor = vectorizer.fit_transform(humor_texts)
y_train_humor = humor_tags
```

```python
X_unsampled_humor, y_unsampled_humor = X_train_humor,y_train_humor
pd.Series(y_unsampled_humor).value_counts()
```

```python
X_test_humor = vectorizer.transform(list(df_humor_val['Sentence']))
y_test_humor = list(df_humor_val['Tag'])
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
def generate_classification_report(model_name, y_test, y_pred):
    report = classification_report(y_test, y_pred)
    print(f'Classification Report for {model_name}: \n{report}\n')
```

```python

models = [LogisticRegression(), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier()]

for model in models:
  model.fit(X_unsampled_humor, y_unsampled_humor)
  y_pred_humor = model.predict(X_test_humor)
  generate_classification_report(model, y_test_humor, y_pred_humor)
```

```python
# We are gettiing highest f1 score for class 1 by not doing any sampling and using Random Foresst Classifier of 78%
```
