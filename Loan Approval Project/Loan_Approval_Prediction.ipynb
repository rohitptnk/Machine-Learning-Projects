```python
# we use over sampling to balance the data set.
from imblearn.over_sampling import RandomOverSampler

oversample = RandomOverSampler(random_state=42)

X_resampled, y_resampled = oversample.fit_resample(X,y)
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns),pd.Series(y_resampled, name='Loan_Status')], axis=1)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Splitting test and train sets
X_resampled_train, X_resampled_test, y_resampled_train, y_resampled_test = train_test_split(X_resampled, y_resampled, test_size=0.25, random_state=42)
```

```python
# Accuracy Score of Random Forest Classifier

model1 = RandomForestClassifier()
model1.fit(X_resampled_train, y_resampled_train)
y_pred = model1.predict(X_resampled_test)
accuracy = accuracy_score(y_resampled_test, y_pred)*100
print(f'Accuracy Score for Random Forest Classifier: {accuracy:.2f}%')
```

```python
# Cross Validation Scores for Logistic Regression

print(f'Cross Validation Score for Logistic Regression: {np.mean(cross_val_score(LogisticRegression(), X, y, cv=5))*100:.2f}%')
```

```python
# Defining a classification report function

from sklearn.metrics import classification_report

def generate_classification_report(model_name, y_test, y_pred):
    report = classification_report(y_test, y_pred)
    print(f'Classification Report for {model_name}: \n{report}\n')
```

```python
# Generating classification report for all the models

models = [LogisticRegression(), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier()]

for model in models:
    model.fit(X_resampled_train, y_resampled_train)
    y_pred = model.predict(X_resampled_test)
    generate_classification_report(model, y_resampled_test, y_pred)
```
